```python
# load tweets text file
text = open("/Users/Labyrinths/Desktop/tweet_text_oct2020IPD.txt", encoding="utf-8").read()
```

```python
# remove punctuation and make lowercase 
def text_to_words(text):
    '''Transform a string to a list of words,
    removing all punctuation.'''
    text = text.lower()

    p = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
    text = ''.join([ch for ch in text if ch not in p])

    return text.split()
```

```python
# returns list of tokenized words from tweets text file
text_to_words(text)
```

```python
# create new list where text words intersect with affect words
def count_occurences(text, affect_word_list):
    '''Count occurences of words from a list in a text string.'''
    text_list = text_to_words(text)

    intersection = [w for w in text_list if w in affect_words_list]

    return len(intersection)
```

```python
count_occurences(text, affect_words_list)
```
